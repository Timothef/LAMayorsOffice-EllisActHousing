```python
import pandas as pd
```

```python
df = pd.read_csv('./data/modified_data/la_housing_dataset_no_geo_property_id_cleanzip2.csv', 
                 encoding='latin-1', index_col=[0])
```

```python
df.columns
```

```python
# CHANDLER'S SCRIPT

#Property's on the RSO list
RSO = df[df['General Category'] == "Is in RSO Inventory"]['Property ID'].unique()

#Property's that have applied for an Ellis Withdrawal
ellis = df[df['General Category'] == "Ellis Withdrawal"]['Property ID'].unique()

#set of all property's that are either on the RSO list or withdrew for Ellise
prewarn = set(RSO).union(ellis)

#create a new dataframe of only property IDs for early warning prediction
train_df= df[df['Property ID'].isin(prewarn)]
```

```python
filter_df = df[df['Property ID'].isin(train_df['Property ID'].unique())]
```

```python
filter_df = df[df['Property ID'].isin(train_df['Property ID'].unique())]
filter_df['Work Description'] = filter_df['Work Description'] .fillna(" ")
agg_work_description = filter_df.groupby(['Property ID'])['Work Description']\
                         .apply(list)\
                         .apply(lambda x: ' '.join(x))   
```

```python
pd.DataFrame(agg_work_description).reset_index().to_csv("./data/modified_data/agg_work-description.csv", encoding='utf-8')
```
